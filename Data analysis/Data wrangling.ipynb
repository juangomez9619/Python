<h1> Transformación de los datos</h1>

<h2> Importar los datos </h2>

```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
df = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv",
                names = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"])
```

```python
df.head()
```

   <div>
    Pasos:
    <ol>
        <li>Identificar los datos perdidos</li>

<h3>Convertir "?" en NaN </h3>

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.dtypes
```

<h3> Identificando valores pérdidos</h3>
<p> el método <code> .isnull()</code> permite encontrar los valores pérdidos (False) para poder ser reemplazados 
    más facilmente </p>

```python
missing_data = df.isnull() 
# también existe .notnull()
missing_data.tail()# checking missing values 
# True -> missing value NaN
# False -> not a missing value
```

```python
columns = missing_data.columns.values.tolist() # lista con los encabezados de las columnas
print(columns)
```

```python
for column in columns:
    print(column)
    print(missing_data[column].value_counts()) # counts True and False values
    print('----------')
```

<h3> Resolver el problema de valores perdidos </h3>
 <ol>
    <li>Eliminar el dato:<br>
        a. Eliminar todo el registro. (fila)
        b. Eliminar la columna.
    </li>
    <li>
        Reemplazar el dato:<br>
        a. Reemplazar por el promedio. <br>
        b. Reemplazar por el dato de mayor frecuecia<br>
        c. Reemplazarlo en base a otras funciones
    </li>
 </ol>   

<h3>Reemplazar por el promedio </h3>

```python

avg_norm_loss = df['normalized-losses'].astype("float").mean(axis = 0)
print(f"avg_norm_loss: {avg_norm_loss}")
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
avg_bore = df['bore'].astype("float").mean(axis = 0)
print(avg_bore)
df['bore'].replace(np.nan, avg_bore, inplace = True)
```

<h3>Reemplazar por dato de mayor frecuencia </h3>

```python
df['num-of-doors'].value_counts().idxmax() # most frequent value 
```

```python
df['num-of-doors'].replace(np.nan,df['num-of-doors'].value_counts().idxmax(), inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
# change data type
df.dtypes
```

```python
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
# Convert mpg to L/100km
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
df.columns.values.tolist()
```

```python
# normalize 'stroke' column
df['stroke'] = df['stroke'].astype('float')
df['stroke'] = df['stroke']/df['stroke'].max() 
```
